```python
# Mengimpor library yang diperlukan
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import bm3d
```

# Gunakan gunakan algoritma Non Local Means untuk denoising: Gaussian, Speckle dan S&P
### Fungsi: ***cv2.fastNlMeansDenoisingColored(a, b, c, d, e, f)***

*   a = image yang akan didenoise
*   b = estimasi tingkat keparahan noise pada image
*   c = tingkat kekuatan denoising
*   d = ukuran template size image yang akan didenoising
*   e = ukuran pencarian pola berulang dalam image
*   f = ukuran intensitas cahaya








```python
# Memuat gambar
image = cv2.imread('Gaussian.jpg')
# Menampilkan gambar asli
cv2_imshow(image)

# Memfilter noise Gaussian dengan Non-Local Means Denoising
gaussian_denoised = cv2.fastNlMeansDenoisingColored(image, None, 7, 10, 7, 21)
cv2_imshow(gaussian_denoised)
```

```python
# Memuat gambar
image = cv2.imread('Speckle.jpg')
# Menampilkan gambar asli
cv2_imshow(image)


# Memfilter noise Speckle dengan Non-Local Means Denoising
speckle_denoised = cv2.fastNlMeansDenoisingColored(image, None, 5, 10, 7, 15)
cv2_imshow(speckle_denoised)
```

```python
# Memuat gambar
image = cv2.imread('s&p.jpg')
# Menampilkan gambar asli
cv2_imshow(image)

# Memfilter noise Salt & Pepper dengan Non-Local Means Denoising dan filter median
sp_denoised = cv2.fastNlMeansDenoisingColored(image, None, 9, 10, 7, 15)
sp_denoised = cv2.medianBlur(sp_denoised, 1)
cv2_imshow(sp_denoised)
```

```python
# Memuat gambar
image = cv2.imread('Poisson.jpg')
# Menampilkan gambar asli
cv2_imshow(image)


# Memfilter noise Poisson dengan library bm3d
sigma_psd = 20  # Atur nilai sigma_psd sesuai kebutuhan
poisson_denoised = bm3d.bm3d(image, sigma_psd)
cv2_imshow(poisson_denoised)
```
